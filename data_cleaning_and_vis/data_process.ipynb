```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
crime_data = pd.read_csv("c:/Users/idowu/Desktop/Machine Learning/crime_ML/crime-analysis/Crime_Data_from_2020_to_Present.csv")
```

```python
crime_data.head()
```

```python
crime_data["Crm Cd Desc"].unique()
```

```python
#select columns needed
crime_data = crime_data[["Date Rptd", "Crm Cd", "Crm Cd Desc"]]
```

```python
#let see the most frequent crimes
crime_data["Crm Cd Desc"].value_counts().head(10)
```

```python
#sort data by crm code
crime_data = crime_data.sort_values(by="Crm Cd")
```

```python
#how many unique crimes do we have
crime_data["Crm Cd"].nunique()
```

```python
#import train_test_split
from sklearn.model_selection import train_test_split
train_data,test_data = train_test_split(crime_data, test_size=0.3, random_state=42)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

#encode the target variable
le = LabelEncoder()
train_data["Crm Cd_Converted"] = le.fit_transform(train_data["Crm Cd"])

```

```python
from sklearn.feature_extraction.text import CountVectorizer

#define the pipeline
pipeline = Pipeline([
    ("vectorizer", CountVectorizer(stop_words="english")),
    ("classifier", RandomForestClassifier(n_estimators=100, random_state=42))
])

# Fit the model
pipeline.fit(train_data["Crm Cd Desc"], train_data["Crm Cd_Converted"])
```

```python
# Make predictions
predictions = pipeline.predict(test_data["Crm Cd Desc"])

```

```python
#inverse transform the predictions (just checkingg)
test_data["Crm Cd"] = le.inverse_transform(test_data["Crm Cd_Converted"])
```

```python


#classification report
print(classification_report(test_data["Crm Cd_Converted"], predictions, target_names=test_data["Crm Cd"].unique()))
```

```python
from sklearn.model_selection import cross_val_score, KFold

#cross validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

print(f"Cross validation score: {cross_val_score(pipeline, train_data['Crm Cd Desc'], train_data['Crm Cd_Converted'], cv=cv, scoring='accuracy').mean()}")
#print scores 
```
